```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re  # puncation erform karna hota h tab use karte h 
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_word=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # lower case 
    text=str(text).lower()
    # removed puncatuations
    text=re.sub(r'[^a-zA-Z\s]','',text)
    words=text.split()
    # removed stopwords
    words=[word for word in words if word not in stop_word]

    # lemmatizer
    words=[lemmatizer.lemmatize(word) for word in words]

    return " ".join(words)
```

```python
df['tags']=df['tags'].apply(preprocess_text)
```

```python
df['tags'][1]
```

```python
df=df.reset_index(drop=True)
```

**words into Vector**

```python
indices=pd.Series(df.index,index=df['title']).drop_duplicates()
```

```python
indices.to_frame()
```

**TFIDF - NLP**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix=tfidf.fit_transform(df['tags']) # it convert 50000* 45447
```

```python
tfidf_matrix
```

**Cosine Similrity**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10):  # title coulmns ka naam h
    if title not in indices:
        return ['Movie Not found']
    
    idx=indices[title]
    sim_score=cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_idx=sim_score.argsort()[::-1][1:n+1]# argsort is sort the basis of indices
    return df['title'].iloc[similar_idx]
```

```python
recommend('Avenger',2)
```

```python
## Pickle Files

import pickle
pickle.dump(tfidf_matrix,open('tfidf_matrix.pkl','wb'))
pickle.dump(indices,open('indices.pkl','wb'))

df.to_pickle('df.pkl')

pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
